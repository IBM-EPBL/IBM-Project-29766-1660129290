# Outlier Detection

```python
df.shape
```

```python
sns.boxplot(df.Height)
```

```python
q1=df.Height.quantile(0.25)  
q3=df.Height.quantile(0.75)
```

```python
IQR=q3-q1
```

```python
upper_limit= q3 + 1.5*IQR

lower_limit= q1 - 1.5*IQR
```

```python
upper_limit
```

```python
lower_limit
```

```python
df.median()
```

```python
df['Height']= np.where(df['Height']<lower_limit,0.039999999999999994 ,df['Height'])
```

```python
sns.boxplot(df.Height)
```

# Split the Data into Dependent and Independent variables.

```python
x=df.drop(columns=['Age'],axis=1)
x
```

```python
y
```

# Scale the independent variables


```python
from sklearn.preprocessing import scale
```

```python
x_scaled=pd.DataFrame(scale(x),columns=x.columns)
x_scaled.head()
```

# Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

# Model building

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression() 
```

```python
model.fit(X_train,y_train)
```

# Testing

```python
pred_test=model.predict(X_test)
```

```python
pred_test
```

```python
pred_train = model.predict(X_train)
pred_train
```

```python
Age= pd.DataFrame({'Actual_Age':y_test,'pred_Age':pred_test})
Age
```

# Performance using Metrics

```python
from sklearn import metrics 
```

```python
# r2score
```

```python
print(metrics.r2_score(y_test,pred_test)) # test accuracy
```
